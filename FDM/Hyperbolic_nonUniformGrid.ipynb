```python
#Incluímos las librerías que vamos a usar
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def RG( stretch, L, N):
    grid = np.zeros(N+1)
    tanGamma = np.tanh(stretch)

    for i in range(0, N+1):
        epsilon = 1 - 2*(i)/(N)
        grid[i] = (1 - np.tanh(stretch*epsilon)/tanGamma)*0.5*L

    return grid

grid = RG(1.2, 1.5, 20)

plt.plot(grid, 'o')
```
